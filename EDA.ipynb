```python
import pandas as pd
import json
from pandas.io.json import json_normalize
pd.options.display.max_columns = None
pd.options.display.max_rows = None
```

```python
df = pd.DataFrame()

with open('foodscore/foodscore/food_details.jl') as f:
    for i,line in enumerate(f):
        product = json.loads(line)
#         print(product)
        df = df.append(json_normalize(product))
        if i == 10:
            break
df
```

```python
df = df.reset_index(drop=True)
df
```

```python
import urllib3
from bs4 import BeautifulSoup
http = urllib3.PoolManager()
```

```python
def get_amazon_asn(row):
    url = 'https://www.upcitemdb.com/upc/'
    upc = row['product_url'][40:52]
    url += upc
    r = http.request('GET', url)
    d = {}
    if r.status == 200:
        soup = BeautifulSoup(r.data)
        table = soup.find('table', class_='detail-list')
        data = iter(table.find_all('td'))
        for k in data:
        #     print(k.get_text(), next(data).get_text())
            d[k.get_text().strip()] = next(data).get_text().strip()
#     print(d)
    return pd.Series([d.get('Amazon ASIN:',None), d.get('Last Scanned:',None)])
```

```python
df[['Amazon_ASIN', 'Last_Scanned']] = df.apply(get_amazon_asn, axis=1)
df
```
